## Q3 — Tiny Recommendation System

Each user's taste is represented as a vector of 4 scores across 4 genres — **action, romance, thriller, comedy** — all between 0 and 1.

```
User A: [0.9, 0.1, 0.8, 0.2]
User B: [0.1, 0.9, 0.2, 0.8]
User C: [0.8, 0.2, 0.9, 0.1]
User D: [0.5, 0.5, 0.5, 0.5]
```

### Q3 Part (i) — Similarity Matrix

Compute the **full similarity matrix** — every user's similarity score against every other user — using only a **single NumPy operation** (no loops).

What linear algebra concept makes this possible in one line?

````python
# Input — do not modify
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])
````

````python
# Write your code here:
sim_mat = np.dot(users,users.T)/ (np.linalg.norm(users,axis = 1,keepdims = True) @ np.linalg.norm(users,axis = 1,keepdims = True).T)
sim_mat
````

**What linear algebra concept makes this possible in one line?**

### Q3 Part (iii) — The Problem with Raw Dot Products

A senior engineer tells you:
> *"Raw dot products are a bad measure of similarity — you need to normalize."\*

1. What is the **problem** with raw dot products?
2. Construct a **specific example using NumPy** that demonstrates the flaw.
3. Fix it.

> **Hint:** Take a user with scores `[1.8, 0.2, 1.6, 0.4]`

*Write your answer here (double-click to edit this cell)*

1. Larger vectors produce larger dot product , regardless of whether the users actually share the same taste.So it is magnitude biased.

````python
# Write your code here:
#user A,B,c for marvel or DC movie watching hours
arr = np.array([[10,0],[1,0],[5,5]])

sim_mat_dot = np.dot(arr,arr.T)

print("A and B have identical taste (marvel only)")
print("similarity using raw dot product of A,C =",sim_mat_dot[2,0],"and B,C =",sim_mat_dot[2,1])
print('similarity of preference of A and B w.r.t C are same?',sim_mat_dot[2,0]==sim_mat_dot[2,1])
print("This above is wrong")

sim_cos = np.dot(arr,arr.T)/ (np.linalg.norm(arr,axis = 1,keepdims = True) @ np.linalg.norm(arr,axis = 1,keepdims = True).T)

print("\nsimilarity using cosine of A,C =",sim_cos[2,0],"and B,C =",sim_cos[2,1])
print('similarity of preference of A and B w.r.t C are same?',sim_cos[2,0]==sim_cos[2,1])
print("This above is correct")
````
